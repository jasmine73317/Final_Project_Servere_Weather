```python
temperature_df = pd.read_csv('../Resources/temperature.csv')
print('Temperature:')
print(temperature_df.head())
pressure_df = pd.read_csv('../Resources/pressure.csv')
print('Pressure:')
print(pressure_df.head())
humidity_df = pd.read_csv('../Resources/humidity.csv')
print('Humidity:')
print(humidity_df.head())
wind_speed_df = pd.read_csv('../Resources/wind_speed.csv')
print('Wind Speed:')
print(wind_speed_df.head())
wind_direction_df = pd.read_csv('../Resources/wind_direction.csv')
print('Wind Direction:')
print(wind_direction_df.head())
weather_description_df = pd.read_csv('../Resources/weather_description.csv')
print('Weather Description:')
print(weather_description_df.head())
city_attributes_df = pd.read_csv('../Resources/city_attributes.csv')
print('City Attributes:')
print(city_attributes_df.head())
```

```python
list_of_dfs = [temperature_df, pressure_df, humidity_df, wind_speed_df, wind_direction_df, weather_description_df]
```

```python
[print(f) for f in os.listdir('../Resources')]
```

```python
for df in list_of_dfs:
    df['datetime'] = pd.to_datetime(df['datetime'])
    df.drop(columns = ['Tel Aviv District', 'Eilat','Haifa','Nahariyya','Jerusalem','Montreal','Beersheba','Vancouver','Toronto'], inplace = True)
pressure_df.head()
```

```python
temperature_df.dtypes
```

```python
list_of_cities = []
for col in temperature_df.columns:
    if col == 'datetime':
        continue
    list_of_cities.append(col)
print(list_of_cities)
```

```python
city_dfs = {}
for city in list_of_cities:
    newcity = ''.join(city.split(' ')) + '_df'
    city_dfs[newcity] = pd.DataFrame()
    city_dfs[newcity]['Date and Time'] = temperature_df['datetime']
    city_dfs[newcity]['Temperature'] = temperature_df[city]
    city_dfs[newcity]['Humidity'] = humidity_df[city]
    city_dfs[newcity]['Wind_Speed'] = wind_speed_df[city]
    city_dfs[newcity]['Wind_Direction'] = wind_direction_df[city]
    city_dfs[newcity]['Pressure'] = pressure_df[city]
    city_dfs[newcity]['Weather_Description'] = weather_description_df[city]
    city_dfs[newcity].dropna(how='any',inplace=True)
    
for df in city_dfs:
#     df['Date and Time'] = pd.to_datetime(df['Date and Time'])
    city_dfs[df]['Year'] = city_dfs[df]['Date and Time'].dt.year
    city_dfs[df]['Month'] = city_dfs[df]['Date and Time'].dt.month
    city_dfs[df]['Day'] = city_dfs[df]['Date and Time'].dt.day
    city_dfs[df]['Hour'] = city_dfs[df]['Date and Time'].dt.hour
    
    
#     df[''] = df.groupby('indicator')['value'].transform('mean')
    city_dfs[df]['Year Temp Max'] =  city_dfs[df].groupby('Year')['Temperature'].transform('max')#get max
    city_dfs[df]['Year Temp Min'] = city_dfs[df].groupby('Year')['Temperature'].transform('min') #get min
    city_dfs[df]['Year Temp Average'] = city_dfs[df].groupby('Year')['Temperature'].transform('mean')
    city_dfs[df]['Month Temp Max'] = city_dfs[df].groupby('Month')['Temperature'].transform('max') #get max
    city_dfs[df]['Month Temp Min'] = city_dfs[df].groupby('Month')['Temperature'].transform('min') #get min
    city_dfs[df]['Month Temp Average'] = city_dfs[df].groupby('Month')['Temperature'].transform('mean') #get average
    
    city_dfs[df]['Year Pressure Max'] = city_dfs[df].groupby('Year')['Pressure'].transform('max') #get max
    city_dfs[df]['Year Pressure Min'] = city_dfs[df].groupby('Year')['Pressure'].transform('min') #get min
    city_dfs[df]['Year Pressure Average'] = city_dfs[df].groupby('Year')['Pressure'].transform('mean')
    city_dfs[df]['Month Pressure Max'] = city_dfs[df].groupby('Month')['Pressure'].transform('max') #get max
    city_dfs[df]['Month Pressure Min'] = city_dfs[df].groupby('Month')['Pressure'].transform('min') #get min
    city_dfs[df]['Month Pressure Average'] = city_dfs[df].groupby('Month')['Pressure'].transform('mean') #get average
    
    city_dfs[df]['Year Wind_Speed Max'] = city_dfs[df].groupby('Year')['Wind_Speed'].transform('max') #get max
    city_dfs[df]['Year Wind_Speed Min'] = city_dfs[df].groupby('Year')['Wind_Speed'].transform('min') #get min
    city_dfs[df]['Year Wind_Speed Average'] = city_dfs[df].groupby('Year')['Wind_Speed'].transform('mean')
    city_dfs[df]['Month Wind_Speed Max'] = city_dfs[df].groupby('Month')['Wind_Speed'].transform('max') #get max
    city_dfs[df]['Month Wind_Speed Min'] = city_dfs[df].groupby('Month')['Wind_Speed'].transform('min') #get min
    city_dfs[df]['Month Wind_Speed Average'] = city_dfs[df].groupby('Month')['Wind_Speed'].transform('mean') #get average
    
    city_dfs[df]['Year Humidity Max'] = city_dfs[df].groupby('Year')['Humidity'].transform('max') #get max
    city_dfs[df]['Year Humidity Min'] = city_dfs[df].groupby('Year')['Humidity'].transform('min') #get min
    city_dfs[df]['Year Humidity Average'] = city_dfs[df].groupby('Year')['Humidity'].transform('mean')
    city_dfs[df]['Month Humidity Max'] = city_dfs[df].groupby('Month')['Humidity'].transform('max') #get max
    city_dfs[df]['Month Humidity Min'] = city_dfs[df].groupby('Month')['Humidity'].transform('min') #get min
    city_dfs[df]['Month Humidity Average'] = city_dfs[df].groupby('Month')['Humidity'].transform('mean') #get average
    
    
city_dfs['SanFrancisco_df'].head()
```

```python
for df in city_dfs:
    city_dfs[df].to_csv(f'../Resources/finished_city_dfs/{df}.csv')
```
