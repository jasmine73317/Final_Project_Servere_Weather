```python
import pandas as pd
import os
```

```python
city_dfs = {}
for f in os.listdir('../Resources/finished_city_dfs'):
    city_dfs[f.split('.')[0]] = pd.read_csv(f'../Resources/finished_city_dfs/{f}')
```

```python
for city in city_dfs:
    print(city)
```

```python
city_dfs['SanFrancisco_df'].head()
```

```python
short_dfs = {}
for city in city_dfs:
    short_dfs[city] = city_dfs[city].loc[city_dfs[city]['Year'] == 2016,:]
```

```python
for df in short_dfs:
    short_dfs[df].to_csv(f'../Resources/limit_one_year_csv/2016_{df}.csv')
```
