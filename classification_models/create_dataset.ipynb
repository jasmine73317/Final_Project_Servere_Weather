```python
df = df.merge(cities_df, how="left", on="City")
df.head()
```

```python
date_time = []
for x in range(0,len(df)):
    date = datetime.datetime.strptime(df["Date and Time"][x], "%Y-%m-%d %H:%M:%S")
    date_time.append(date)
df["Timestamp"] = date_time
df["Timestamp"] = df["Timestamp"].map(datetime.datetime.timestamp)
```

```python
day = 24*60*60
year = (365.2422)*day

timestamp = df["Timestamp"]

df["Day sin"] = np.sin(timestamp * (2*np.pi/day))
df["Day cos"] = np.cos(timestamp * (2*np.pi/day))
df["Year sin"] = np.sin(timestamp * (2*np.pi/year))
df["Year cos"] = np.cos(timestamp * (2*np.pi/year))
```

```python
df = df.loc[:,['Date and Time', 'Timestamp', 'Temperature', 'Humidity', 'Wind_Speed',
        'Wind_Direction', 'Pressure', 'Weather_Description', 'Year',
        'Day sin', 'Day cos', 'Year sin', 'Year cos',
        'City', 'Latitude', 'Longitude']]
df = df.sort_values(by=["Timestamp", "City"])
df.head()
```
